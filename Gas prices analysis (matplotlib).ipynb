## Read world analysis

### Line graph

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
gas.columns
```

```python
plt.figure(figsize=(8,5))

plt.plot(gas.Year, gas.USA, 'r.-', label="USA")           # r = red color, .(dot) = is for marking points on the line.
plt.plot(gas.Year, gas.Canada, 'b.-', label="Canada")     # b = blue color.

plt.title('USA vs Canada gas prices')
plt.ylabel('Dollars/Gallon')
plt.xlabel('Year')
plt.legend()

plt.xticks(gas.Year[::3].tolist()+[2011])  # year[::3] is for to set the interval to 3 years   
plt.savefig('Gas prices.png', dpi=150)
plt.show()                                 # and tolist is for setting the last value (like tolerance)
```

```python
for country in gas:
    if country != 'Year':                             # how to move the legend out of the graph? 
        plt.plot(gas.Year, gas[country], marker='.')
        plt.xticks(gas.Year[::3].tolist()+[2011])
```

```python
countries = ['Australia', 'USA', 'Canada', 'South Korea']
for country in gas:                                        # another way of plotting.
    if country in countries:
        plt.plot(gas.Year, gas[country], marker='.')
        plt.xticks(gas.Year[::3].tolist()+[2011])
```
